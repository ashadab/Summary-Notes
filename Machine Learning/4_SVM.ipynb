Hypothesis of Logistic regression
$h_{\theta}(x) =p(\hat{y}=1|x) $

**Prediction:** 

$\hat{y} =1$ (Class 1) if $ h_{\theta}(x) >= T $
            
$\hat{y} =0$ (Class 0) if $ h_{\theta}(x) < T $

where $T$ is the user-defined threshold

SVM is another supervised learning algorithm used for classification problem. It is a non-probabilistic classifier.

### Derviation of SVM Cost function:

We start with Logistic regressin cost function and then try to modify it.

$\begin{align*}J(\theta) & = \frac{1}{m}\sum_{i=1}^m -y^{(i)} \log(h_\theta(x^{(i)})) - (1 - y^{(i)})\log(1 - h_\theta(x^{(i)}))\\ & = \frac{1}{m}\sum_{i=1}^m -y^{(i)} \log\Big(\dfrac{1}{1 + e^{-\theta^Tx^{(i)}}}\Big) - (1 - y^{(i)})\log\Big(1 - \dfrac{1}{1 + e^{-\theta^Tx^{(i)}}}\Big)\end{align*} $

let $ z= \theta^{T} *x $, we now approximate the logistic regression curves

```python
z=np.linspace(-4,4)

#
y_1= - np.log( 1 / ( 1+ np.exp(-x) ) )

y_0= - np.log( 1 - ( 1 / ( 1+ np.exp(-x) ) )   )

#plt.figure(figsize=(20,20))
fig, axp = plt.subplots(1,2, figsize=(10,5))

axp[0].plot(z,y_1,'b', label='Cost function when y=1')
axp[0].legend()
axp[0].set_title('Cost_1')
axp[0].set_xlabel('z')

axp[1].plot(z,y_0,'r', label='Cost function when y=0')
axp[1].legend()
axp[1].set_title('Cost_0')
axp[1].set_xlabel('z')

print( 'Value of cost function in both the graphs at z=0 is ', - np.log(0.5) )
```

let $ z= \theta^{T}x $, we now approximate the logistic regression curves.

Section of cost function when y=1:

$\text{cost}_1(z) = \max(0, k(1-z))$ where $k>0$ is any arbitrary constant. Note that cost shoots up when $z << 0 $.

<br>

Section of cost function when y=0:

$\text{cost}_0(z) = \max(0, k(1+z))$ where $k>0$ is any arbitrary constant. Note that cost shoots up when $z >> 0 $.





$J(\theta) = \frac{1}{m} \sum_{i=1}^m y^{(i)}(-\log(h_\theta(x^{(i)}))) + (1 - y^{(i)})(-\log(1 - h_\theta(x^{(i)}))) + \dfrac{\lambda}{2m}\sum_{j=1}^n \Theta^2_j $

$J\theta) = \frac{1}{m} \sum_{i=1}^m y^{(i)} \ \text{cost}_1(\theta^Tx^{(i)}) + (1 - y^{(i)}) \ \text{cost}_0(\theta^Tx^{(i)}) + \dfrac{\lambda}{2m}\sum_{j=1}^n \Theta^2_j$

$J(\theta) = \sum_{i=1}^m y^{(i)} \ \text{cost}_1(\theta^Tx^{(i)}) + (1 - y^{(i)}) \ \text{cost}_0(\theta^Tx^{(i)}) + \dfrac{\lambda}{2}\sum_{j=1}^n \Theta^2_j$

<br> 

**Cost function for SVM:** 

$J(\theta) = C\sum_{i=1}^m y^{(i)} \ \text{cost}_1(\theta^Tx^{(i)}) + (1 - y^{(i)}) \ \text{cost}_0(\theta^Tx^{(i)}) + \dfrac{1}{2}\sum_{j=1}^n \Theta^2_j$

where $ C =  \frac{1}{\lambda} $  
<br>

#### Hinge Loss

Note that above cost/loss function for SVM is also called as Hinge Loss function. It is a convex function, whose general form is:

$l(y) = max(0, 1-t . y) $ where t= +- 1 and $ y= \textbf{w}.\textbf{x} + b $ where $(\textbf{w},b)$ are weight and bias parameter respectively.


Ref: https://en.wikipedia.org/wiki/Hinge_loss

<br> 
**Hypothesis function for SVM:**

$h_\theta(x) =\begin{cases}    1 & \text{if} \ \Theta^Tx \geq 0 \\    0 & \text{otherwise}\end{cases}$

it is because we can easily see that at $ z=\theta^{T}x= 0 $ , $cost= 0.69 $ so the cost is very small when z > 0.

Also, it means that SVM doesn't produce probabilites of an observation lying in Class 1 or 0.

## Large margin classifier

SVM with a large C is also known as Large Margin Classifier. 

In case we have outliers in the data and still use a large value of C. This would lead to overfitting of the data and SVM will no longer be large margin classifer. 

In order to avoid overfitting, we need to reduce C where ($ C = \frac{1}{\lambda} $ )

if C is large, optimization problem becomes following: 

$ \begin{align*}
J(\theta) = C \cdot 0 + \dfrac{1}{2}\sum_{j=1}^n \Theta^2_j \newline
= \dfrac{1}{2}\sum_{j=1}^n \Theta^2_j
\end{align*} $

st. $ \Theta^Tx \geq 1 $ if y=1
<br>
$  \Theta^Tx < -1 $ if y=0

## General Proof for SVM
<br>
For rigorous proof look at below links. In all these proofs, NO regularization is assumed that means C is very large. Also, 2 classes are treated as +1 and -1 rather than 1 and 0. We assume maximum margin/seperating/optimal hyperplane as $w^{T}x + b =0 $ and two of support/margin planes as $w^{T}x + b = \pm1 $ (even if we used $p$ in instead of 1 in RHS, we could divide the entire equation by $p$ to get 1 on RHS).

We get to know that the distance between 2 margins is $\frac{2}{||w||}$. Hence, the optimization reduces to maximization of the distance subject to constraints. 

$ min ||w|| \\ s.t. \\ w^{T}x + b \geq 1 \  \text{if} \  y_{i} =1 \\ w^{T}x + b < -1 \ \text{if} \ y_{i} = -1 $

Note that for a point for which $w^{T}x >> 1$ we would like to predict class 1 and for points like $w^{T}x << -1 $ we would like to predict class -1

**It is important to note that only support vectors (points which are closest to maximum margin hyperplane) determines maximum margin hyperplane. Other points in training set are not consequnetial.**

## Non-linear SVM  (Kernel trick is used)

We use non-linear SVM when the data is not linearly seperable. 

The main idea is to use Kernel function (also known as similarity functions) to create new features which can be used to build the hypothesis function.

Read following:

- Summary of the Kernel https://www.coursera.org/learn/machine-learning/resources/Es9Qo
- See lecture slides/screenshots

Summary:

1. Choose landmarks for the training data set (generally we choose all the m-examples of the training set as the landmark

2. We then construct features based on the Kernel/similarity function for each landmark. 

3. Finally, we combine all the features to find hypothesis function and predict class 1 if theta^T*x > 0

## Big idea 

Main idea of Kernel trick is to transform the feature space from $ R^{n} $ to $ R^{m} $. After doing this, we are able to make the data linearly seperable.


### Cross-validation in SVM

We need to Cross-validate (choose/tune) following while using SVM:

1. $C$ parameter where $ C = \frac{1}{\lambda}$
2. Choice of Kernel
3. If we use Gaussian kerneal than $\sigma^{2}$ needs to be tuned

To address overfitting:

- Reduce C
- Increase $\sigma^{2}$ since if we have large $\sigma^{2}$ then we will have smoother value of feature. This will ensure smoother predictions. Hence, the variance will be lower.

Linear Kernel is same as NO Kernel. Based on magnitude of $ n$ (#features) and $m$ (# training examples) we need to decide what whether to use Logistic reg/SVM without kernel/SVM with Kernel. 

Also, note that SVM without Kernel is almost similar to Logistic regression and hence used interchangebly.

```python
param_grid = { 'C': [0.1,1, 10, 100, 1000] , 'gamma':[1,0.1,0.01,0.001,0.0001], 'kernel': ['rbf']  }

from sklearn.model_selection import GridSearchCV
```
