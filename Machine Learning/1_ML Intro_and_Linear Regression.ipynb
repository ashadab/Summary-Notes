
# ML Introduction

**Supervised Machine Learning**

In Supervised ML, for each example we always receive “true” labels. Supervised ML tasks can be of 2 types: 
1. Regression task: Here the output is real valued
2. Classification task: Here the output is binary (either zero or one)


**Hypothesis**

Idea is to learn a hypothesis (a function) using existing training dataset so that in future ‘hypothesis’ gives us the prediction for the new inputs.


**Cost function**

Cost function is a single number which is average of cumulative differences between the prediction and ground truth for each training example. In order to learn parameters of the hypothesis, we minimize the cost function with respect to parameters.
Idea is to obtain set of parameters which results into minimum possible value of cost function.


**Plot**

In the parameter space, we can plot cost function. In case of linear regression, cost function has happens to be always a convex function.

**Gradient Descent**

If the sample size is very large, it becomes computationally very expensive to find the exact estimates of the hypothesis/function. 
Hence, we turn to a more iterative approach known as gradient descent where we update the parameter at each step. For each j = {0,1,2… n}

$\theta _{j} : = \theta _{j} - \alpha * \frac{\partial J }{\partial \theta_{j}}$ for $ j= {0,1,2...n} $


Note:

Batch gradient takes into account all the examples at each iteration while updating values of a parameter.  
However, Stochastic gradient descent takes into account only a single example while updating the values of parameter. 


**Feature Scaling**

Feature scaling is done in order to achieve either of the following:
1. In algorithms which uses L-2 norm (or Euclidean distance) if one of the features has different range than other, then while computing L-2 norm only one of the feature would dominate.
2. Algorithms which use Gradient descent witness faster convergence if the features happen to be normalized. 

There are 3 major types of feature scaling: 
1. Min-max normalization:  It results into data which is in range [0,1]

$$\frac{x-min(x)}{max(x)-min(x)}$$

2. Mean normalization

$$\frac{x-\mu_{x}}{max(x)-min(x)}$$

3. Standardization

$$\frac{x-\mu_{x}}{ \sigma_{x} } $$

Reference: https://en.wikipedia.org/wiki/Feature_scaling




**Learning Rate ($\alpha$)**

**How to choose right alpha?**

For each $\alpha$, we plot $J(\theta)$ with respect to # iterations. Ideally, at each iteration the cost function should decrease. 

Note: You also need to decide on a stopping criterion when $J(\theta)$ will reduce by a very small number and you would like the algorithm to stop.

Rule of thumb:

1. If $\alpha$ is too small, we will have slow convergence 
2. If $\alpha$ is too large, $J(\theta)$ may not decrease on every iteration; it may not even converge.
Typical values if $\alpha$ are $10^{-3}$ , $10^{-2}$, $10^{-1}$, $1$, … Andrew Ng recommends decreasing α by multiples of 3.

# <font color='voliet'> Linear regression </font>

Multivariate regression model with $n$ regressors/variables and $m$ observations/examples

$ h_{\theta}(x) = \theta^{T}*x $ where $ \theta = \begin{bmatrix} \theta_{0} \\ \theta_{1}  \\ . \\. \\ \theta_{n} \end{bmatrix}$
$ x = \begin{bmatrix} 1 \\ x_{1}  \\ . \\. \\ x_{n} \end{bmatrix} $


$ h_{\theta}(x^{(i)}) = \theta_{0} + \theta_{1}x_{1,i}+ \theta_{2}x_{2,i}.... + \theta_{n}x_{n,i} $ for $i = 1,2, ...m $

Note that we don't include error terms in the model which is unlike the way we used to do in Econometrics because we will just minimize the deviations. In ML, we are just interested in doing predictions after minimizing the loss function and nothing else.

Also, we used a Quadratic Loss function for finding estimates of the model:

$ J(\theta) = \frac{1}{2*m} * \sum_{i=1}^{m} ( y^{(i)} - h_{\theta}(x^{(i)}) )^{2} $

**Goodness of fit ( $ R^{2} $ )**

$ R^{2} = 1 - \frac{SSE}{SST}  = \frac{SSR}{SST} $

where $ SSR = \sum_{i=1}^{m} ( \widehat{ y_{i} } - \bar{y} )^{2} $ , $ SSE =  \sum_{i=1}^{m} ( y_{i} - \bar{y} )^{2} $
and $ SST =  \sum_{i=1}^{m}(y_{i} - \bar{y} )^{2} $

SSR: Square sum of Regression

SSE: Square sum of errors/residuals

SST: Square sum of total error/variation

# Gauss Markov Assumptions of OLS

If Gauss Markov assumptions hold then OLS estimators will be BLUE.

**BLUE estimators : Best ; Linear ; Unbiased ; Efficient**



**Major idea** is that we have a population of X and Y for which we fit a line known as PRF (population regression function) ..i.e. we try to find the relationship between X and Y.

However, in reality, we can never observe/find entire population. So, in order to find relationship between X and Y , we draw a random sample from the population. We then find estimates (best guess) of population parameters using a SRF (sample regression function). 




Gauss Markov assumptions are for the population and NOT the random sample.

**Assumptions**
1. Model is linear in parameters

2. Sample $( X_{i},Y_{i}) $ drawn from the population is RANDOM. i.e. sample is representative of the population.

3. No cross-correlation between errors terms. $ E(u_{i}, u_{j}) = 0 $ for all $ i \neq j $

4. No endogenity. Zero correlation between error terms and explainatory variables i.e $ E(u_{i} | X_{i}) = 0 $ for each $ i = { 1, 2, 3, .. } $

5. No heteroescadisticity. i.e. $ E(u_{i}^{2}) = \sigma^{2}$  for $\forall i $

6. No multi-collinearity. i.e. $ \alpha_{0} + \alpha_{1}X_{1}+ \alpha_{2}X_{2}.... + \alpha_{n}X_{n} \neq 0   $
otherwise the matrix X becomes non-invertible since 2 or more columns will be dependent


We can combine assumption 3 and 5 to say that $u_{i}$ are $iid$ (independently and identically distributed). 

Additional Assumption

7. To make use of finite sample properties like hypothesis testing we can impose an additional restriction that error terms are normally distributed. i.e.  

$ u_{i} \sim N(0,\sigma^{2}) $
