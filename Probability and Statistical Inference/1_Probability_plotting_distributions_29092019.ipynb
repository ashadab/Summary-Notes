# Chi-square distribution

Chi square distribution is obtained by taking sum of squares of k-standard normal random variables.

Degree of freedom = # standard normal rv used = Mean of chi-square random variable

2 methods of finding square of numbers

x_chi = np.random.normal(0,1,10)

#list comprehension
cf= [ x**2 for x in x_chi ] 

#using lambda function
sq = lambda x: x**2


```python
sum_series = []

rand_sample_size = 100000

for iter in range( rand_sample_size  ):
    
    #degree of freedom
    dof = 5
        
    x_chi = np.random.normal( loc=0, scale= 1, size= dof)
    
    #list comprehension
    cf= [ x**2 for x in x_chi ] 
    
    sum_series.append ( np.sum(cf) )

print( 'mean =', np.mean( sum_series    )  )
    
sns.distplot( sum_series  ) 

len(sum_series)
```

```python
#number of trials
n= 100

#probaility of success
p = 0.5

# sample size to be drawn of iid (i.e. how many times the entire experiment was conducted )
rand_sample_size = 100000

#x_bino counts for how many successes occured in n-trails
x_bino = np.random.binomial(n,p, size = rand_sample_size )

print( x_bino[-5:-1] )

sns.distplot( x_bino, kde= False , color='skyblue',
                  hist_kws={"linewidth": 2, 'rwidth': 2 , 'alpha':0.5 ,'edgecolor':'black' } )

print( 'Sample Mean= ', np.mean(x_bino) )
print('Pop mean = ', n*p , '\n')

print('Sample Variance =' , np.var(x_bino ) )
print('Pop variance = ', n*p*(1-p) , '\n')
```

```python
mu = 3
sigma = 80

a=[]
b=[]
c=[]

for i in [ 1 , 2 , 3 ]:
    
    upper = st.norm.cdf( mu + i * sigma , loc = mu  , scale = sigma )
    a.append(  upper  )
    
    lower = st.norm.cdf( mu - i * sigma , loc = mu ,  scale = sigma )
    b.append(  lower )
    
    c.append(  upper - lower ) 


print(c)    
```

# t-distribution

%%latex

t $\sim$ $\frac{Z}{ \sqrt{ \frac{X}{n} } }   $ where Z is the standard normal and X follows a chi-square distribution
with n-degrees of freedom

```python
random_sample_size = 100

#degree of freedom of chi-squre
df = 4


x_stand_norm = np.random.standard_normal(size = random_sample_size)

x_chi_square = np.random.chisquare(df , size = random_sample_size)

numr = x_stand_norm
denr = np.sqrt( x_chi_square /  random_sample_size  )

t_stat =  numr / denr

plt.figure

ax = sns.distplot(t_stat )

print('population mean = ', 0   )
print('Sample mean = ' , np.mean(t_stat) )

print('population variance =',   df / (df-2)  )
print ('Sample variance = ' , np.var(t_stat))
```

%%latex
If degree of freedom (df) $\rightarrow \infty $ then t-distribution converges to a standard normal distribution

```python
random_sample_size = 100000

#degree of freedom of chi-squre
df = 100000

x_stand_norm = np.random.standard_normal(size = random_sample_size)

x_chi_square = np.random.chisquare(df , size = random_sample_size)

numr = x_stand_norm
denr = np.sqrt( x_chi_square /  random_sample_size  )

t_stat =  numr / denr


print('population mean = ', 0   )
print('Sample mean = ' , np.mean(t_stat) )

print('population variance =',   df / (df-2)  )
print ('Sample variance = ' , np.var(t_stat))


plt.figure
ax = sns.distplot(t_stat )
```
